```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import LogisticRegression

from candlestick import candlestick
import plotly.graph_objects as go
```

```python
# Load stock data
df_ind = pd.read_excel('Index/SP500 tickets.xlsx')
df_ind2 = pd.read_csv('Index/Stocks in the SP 500 Index.csv')

df_ind.columns = [col.replace(u'\xa0', u' ') for col in df_ind.columns]
```

```python
# Create stock and industrial data
# Get all ticker price
stocks = df_ind['Symbol'].unique()

# Standardize the column names
cols = ['report_date', 'open', 'high', 'low', 'close', 'volume', 'adjusted_close']

# Stock Price Data for the above 5 stocks
df_price = pd.DataFrame()

for stock in stocks:
    
    # Read the csv
    temp = pd.read_csv(f'Stock Price/{stock}_2022-07-31.csv')
    temp.columns = cols
    temp['ticker'] = stock
    
    # Standardize the stock price (percentage change)
    # Convert all numeric features to percentage change relative to the first numeric record
#     for col in ['open', 'high', 'low', 'close', 'volume', 'adjusted_close']:
#         temp[col] = (temp[col] - temp[col].iloc[0]) / temp[col].iloc[0]
    
    # Append to the stock price data
    df_price = df_price.append(temp)
```

```python
# List out all industries the stock belongs to
inds = list(df_ind[df_ind['Symbol'].isin(stocks)]['GICS Sector'].unique())
inds_df = df_ind[df_ind['GICS Sector'].isin(inds)]
df_ind_idx = pd.DataFrame()

# Stock Price for all tickers belonging to the industries
df_ind_price = pd.DataFrame()
cols = ['report_date', 'open', 'high', 'low', 'close', 'volume', 'adjusted_close']

for _, row in inds_df.iterrows():
    
    ticker = row['Symbol']
    temp = pd.read_csv(f'Stock Price/{ticker}_2022-07-31.csv')
    temp.columns = cols
    temp['ticker'] = ticker
    temp['inds'] = row['GICS Sector']
    
    df_ind_price = df_ind_price.append(temp)
    
# Create Industrial Index, using close price
temp_ind_idx = df_ind_price.groupby(['report_date', 'inds'], as_index = False).agg({
    'open': np.sum,
    'high': np.sum,
    'low': np.sum,
    'close': np.sum,
    'volume': np.sum,
    'adjusted_close': np.sum
    
#     'ticker': lambda x: list(x)
})

# Standardize the stock price (percentage change)
# Convert all numeric features to percentage change relative to the first numeric record
for ind in inds:
    temp = temp_ind_idx[temp_ind_idx['inds'] == ind]
    for col in ['open', 'high', 'low', 'close', 'volume', 'adjusted_close']:
        temp[col] = (temp[col] - temp[col].iloc[0]) / temp[col].iloc[0]
        
    df_ind_idx = df_ind_idx.append(temp)
```

```python
# Find the most common 8 candlestick pattern 
# Ref. Explainable Deep Convolutional Candlestick Learner
# Github repo for candlestick patterns:
# https://github.com/SpiralDevelopment/candlestick-patterns

stocks = df_price['ticker'].unique()
df_cad = pd.DataFrame()

for stock in stocks:
    temp = df_price[df_price['ticker'] == stock]

    # Morning Star (Have)
    temp = candlestick.morning_star(temp, target = 'morning_star')

    # Evening Star (Have)
    temp = candlestick.evening_star(temp, target = 'evening_star')

    # Hammer (Have)
    temp = candlestick.hammer(temp, target = 'hammer')

    # Inverted Hammer (Have)
    temp = candlestick.inverted_hammer(temp, target = 'inverted_hammer')

    # Bullish Engulfing (Have)
    temp = candlestick.bullish_engulfing(temp, target = 'bullish_engulfing')

    # Bearish Engulfing (Have)
    temp = candlestick.bearish_engulfing(temp, target = 'bearish_engulfing')

    # Shooting Star (Have)
    temp = candlestick.shooting_star(temp, target = 'shooting_star')

    # Hanging Man (Have)
    temp = candlestick.hanging_man(temp, target = 'hanging_man')

    candlestick_cols = temp.columns[8:8+8]
    temp['candlestick_label'] = temp.iloc[:,8:8+8]\
        .apply(lambda arr: ','.join([candlestick_cols[ind] for ind, label in enumerate(arr) if label]), axis = 1)
    
    # Standardize the stock price (percentage change)
    # Convert all numeric features to percentage change relative to the first numeric record
    for col in ['open', 'high', 'low', 'close', 'volume', 'adjusted_close']:
        temp[col] = (temp[col] - temp[col].iloc[0]) / temp[col].iloc[0]
    
    df_cad = df_cad.append(temp)
    
df_cad = df_cad.dropna()
```

```python
# Create label
# Using Slop-detection method (using close)
# Reference: Short-Term Stock Price-Trend Prediction Using Meta-Learning
K = 14
df_tar = pd.DataFrame()

for stock in stocks:
    temp = df_price[df_price['ticker'] == stock]
    temp['avg_close_before'] = temp['close'].rolling(K).mean()
    temp['avg_close_after'] = temp['close'].shift(-K).rolling(K).mean()
    temp['slope'] = temp['avg_close_after'] - temp['avg_close_before']
    temp = temp.dropna().reset_index(drop = True)

    # Label the Slope
    temp['label_str'] = temp['slope'].apply(lambda x: 'rise' if x > 0 else 'drop')
    temp['label'] = temp['slope'].apply(lambda x: 1 if x > 0 else 0)
    df_tar = df_tar.append(temp)
```

```python
# Visualize the ts with label
np.random.choice(7600)
stock = np.random.choice(stocks)
print(stock)
temp = df_tar[df_tar['ticker'] == stock]

fig = plt.figure(figsize=(12,8))
temp_rise = temp[temp['label_str'] == 'rise']
temp_drop = temp[temp['label_str'] == 'drop']

plt.plot(temp['close'], label = 'Close')
plt.plot(temp_rise.index, temp_rise['close'], 'go', label = 'Rise')
plt.plot(temp_drop.index, temp_drop['close'], 'ro', label = 'Drop')
plt.legend()
plt.xlabel('Time index')
plt.ylabel('% Change')

fig.show()
```

```python
# Merge the result
df_tar = df_tar[['report_date', 'ticker', 'label']]
data = df_tar.merge(df_cad, on = ['report_date', 'ticker'])

feature_cols = ['open', 'high', 'low', 'close', 
                'morning_star', 'evening_star', 'hammer', 'inverted_hammer', 'bullish_engulfing',
                'bearish_engulfing', 'hanging_man']
target_col = ['label']
```

```python
# Train-test split
test_size = 0.2

report_dates = data['report_date'].unique()
train_report_dates = report_dates[:int(len(report_dates) * (1-test_size))]
test_report_dates = report_dates[int(len(report_dates) * (1-test_size)):]

train = data[data['report_date'].isin(train_report_dates)]
test = data[data['report_date'].isin(test_report_dates)]

X_train = train[feature_cols]
y_train = train[target_col]
X_test = test[feature_cols]
y_test = test[target_col]

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Fit Logistic Regression and calculate the accuracy
clf = LogisticRegression(random_state = 7600)
clf.fit(X_train, y_train)
```

```python
print('Training Accuracy: ', round(clf.score(X_train, y_train) * 100, 2), '%')
print('Testing Accuracy: ', round(clf.score(X_test, y_test) * 100, 2), '%')
```
