```python
# News
df_news = pd.read_csv('news/vader_sentiment.csv')
df_news['report_date'] = df_news['published_utc'].apply(lambda x: datetime.strptime(str(x), '%Y%m%d'))
df_news = df_news.drop('published_utc', axis = 1)
df_news['ticker'] = df_news['ticker'].apply(lambda x: x.replace('.', '-'))

# covid
df_covid = pd.read_csv('covid/covid.csv')
df_covid['date'] = df_covid['date'].apply(lambda x: datetime.strptime(x, "%Y-%m-%d"))

# Natural diaster
df_diaster = pd.read_csv('disaster/disaster.csv')
df_diaster = df_diaster[['date', 'Total Deaths']]
df_diaster['date'] = df_diaster['date'].apply(lambda x: datetime.strptime(str(x), "%Y%m%d"))

# Map to covid count
df_news = df_news.merge(df_covid, left_on = ['report_date'], right_on = ['date'], how = 'left')
df_news = df_news[df_news['date'].isnull() == False].drop('date', axis = 1)

# Map to natural diaster
df_news = df_news.merge(df_diaster, left_on = ['report_date'], right_on = ['date'], how = 'left')
df_news['Total Deaths'] = df_news['Total Deaths'].fillna(0)
df_news = df_news.drop('date', axis = 1)
df_news = df_news.rename({'Total Deaths': 'total_deaths'}, axis = 1)

df_stock_news = df_news.copy()
```
